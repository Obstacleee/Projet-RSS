## 1. Introduction
### 1.1 Qu'est-ce que [RSS](https://fr.wikipedia.org/wiki/RSS) ?
RSS, qui signifie `Really Simple Syndication` ou `Rich Site Summary`, est un type de flux web qui permet aux utilisateurs et aux applications d'accéder à des mises à jour de contenu en ligne dans un format standardisé et lisible par un ordinateur.

  <h4>Logo Flux RSS</h4>
</div>

Ces flux peuvent, par exemple, permettre à un utilisateur de suivre de l'actualité de nombreux sites web. 
Ces flux permettent donc de transmettre des infos d'un site web à un autre ou d'un site web à un utilisateur.

> Cette méthode de distribution de contenu est populaire pour la diffusion d'articles, d'audio et de vidéo.

### 1.2 Importance des RSS
Les flux RSS sont cruciaux pour les personnes qui veulent recevoir des mises à jour en temps opportun de leurs sites web préférés sans avoir à visiter chacun individuellement.

Ils offrent un moyen pratique de rester informé et sont particulièrement utiles pour :

- **Agrégation de contenu** : Le RSS permet aux utilisateurs d'agréger du contenu provenant de sources multiples en un seul endroit.
- **Efficacité temporelle** : Il permet de gagner du temps en apportant les dernières mises à jour directement à l'utilisateur, plutôt que de demander à l'utilisateur de vérifier chaque site manuellement.
- **Flux d'actualités personnalisés** : Les utilisateurs peuvent s'abonner à des flux spécifiques qui les intéressent, créant ainsi un flux personnalisé de nouvelles et d'informations.
- **Automatisation** : Les flux RSS peuvent être analysés et traités par des logiciels pour fournir du contenu de manière innovante, comme déclencher des notifications lorsque de nouveaux contenus sont publiés.

### 1.3 Les avantages de `Flux RSS vers Audio`

**Il s'agit d'une application qui tire parti de la puissance des flux RSS pour fournir du contenu sous forme audio.**

Cela est particulièrement bénéfique pour les utilisateurs qui préfèrent ou qui ont besoin de l'audio plutôt que du texte pour des raisons de commodité, d'accessibilité ou de préférence personnelle.

Voici pourquoi `Flux RSS vers Audio` est un développement significatif :

- **Accessibilité** : Convertit les actualités basées sur du texte en audio, ce qui est essentiel pour les utilisateurs malvoyants ou ceux qui ont des difficultés à lire du texte sur des écrans.
- **Multitâche** : Permet aux utilisateurs d'écouter les dernières nouvelles tout en effectuant d'autres activités, telles que conduire, faire de l'exercice ou travailler.
- **Apprentissage des langues** : Peut être utilisé comme outil pour les apprenants de langues qui veulent améliorer leurs compétences d'écoute.
- **Consommation de contenu** : Offre une manière alternative de consommer du contenu pour les utilisateurs qui trouvent que le contenu audio est plus engageant ou plus facile à comprendre que le texte écrit.

### 1.4 Objectifs 

L'objectif principal est de **développer une application Streamlit qui permet aux utilisateurs d'écouter des fichiers audio générés à partir d'entrées de flux RSS.**

Les objectifs spécifiques associés sont :

1. **Extraire et stocker les flux RSS :** Récupérer les données à partir de flux RSS spécifiés et les stockés par thématiques dans une base de données structurée.

6. **Automatisation des Mises à Jour de Flux :** Établir une routine pour la mise à jour automatique des entrées de flux et la génération de nouveaux fichiers audio.

8. **Déploiement de l'Application :** Déployer l'application Streamlit sur un serveur ou une plateforme de cloud computing, permettant un accès global.


## Overview

The script completes its task in the following steps:

1. **Parse RSS Feed**: The script starts by parsing an RSS feed from a specified URL using the `feedparser` library.
2. **Text-to-Speech Conversion**: Next, it converts the extracted titles from the feed to speech using OpenAI's text-to-speech API.
3. **Merge Audio Files**: After generating the speech audio for each title, the script merges all individual audio files into one single audio file.
4. **Cleanup**: Finally, the script removes all the temporary audio files that were created during the process.

## Detailed Steps

### Step 1: Parse RSS Feed

- The script uses the `feedparser` library to parse the RSS feed from "Le Monde" sports section.
- Only the titles of the feed entries are used for speech conversion in this script, but it can be modified to use summaries if a longer text is desired.

### Step 2: Text-to-Speech Conversion

- It sends a request to OpenAI's text-to-speech API with each title.
- Upon a successful API call, the returned audio content is saved as a temporary MP3 file.
- The path to each temporary MP3 file is stored for later use.

### Step 3: Merge Audio Files

- The script uses the `moviepy.editor` library to concatenate all the temporary MP3 files into a single MP3 file.
- The final merged audio is saved to a predefined path (`final_merged_audio.mp3`).

### Step 4: Cleanup

- The script removes all the temporary MP3 files that were created during the text-to-speech conversion.
- It prints the location of the final merged MP3 file to the console.

## Security Note

The script is designed to retrieve the OpenAI API key from an environment variable rather than hard-coded into the script. This is a best practice for maintaining security and confidentiality of sensitive information.

## Running the Script

To run the script:

1. Ensure you have the `feedparser`, `numpy`, `requests`, `openai`, and `moviepy` libraries installed in your Python environment.
2. Set the `OPENAI_API_KEY` environment variable with your OpenAI API key.
3. Execute the script either through a command-line interface or an IDE.

Upon execution, the script will output the path to the final merged audio file, indicating successful completion of all steps.

```python
# Import necessary libraries
import feedparser  # Library for parsing RSS feeds
import tempfile    # Library for creating temporary files and directories
import numpy as np # Library for numerical operations
import os          # Library for interacting with the operating system
import requests    # Library for making HTTP requests
import openai      # Library for interacting with OpenAI API
from moviepy.editor import concatenate_audioclips, AudioFileClip  # MoviePy library for audio editing

# Set OpenAI API key using environment variables for better security practices
# openai_api_key = os.getenv('OPENAI_API_KEY')
openai_api_key = ""
```

```python
def text_to_speech(text, voice='alloy'):
    """
    Converts the provided text to speech using OpenAI's text-to-speech API.

    Parameters:
    - text (str): The text to be converted to speech.
    - voice (str, optional): The voice model to be used for the text-to-speech conversion. Defaults to 'alloy'.

    Returns:
    - str: The file path to the temporary MP3 file containing the speech.

    Raises:
    - Exception: If the request to OpenAI's API fails.

    """
    response = requests.post(
        "https://api.openai.com/v1/audio/speech",
        headers={"Authorization": f"Bearer {openai_api_key}"},
        json={"model": "tts-1", "input": text, "voice": voice}
    )
    
    # Raise an exception if the API call was unsuccessful
    if response.status_code != 200:
        raise Exception(f"Request failed with status code: {response.status_code}")

    # Save the speech to a temporary MP3 file and return the file path
    with tempfile.NamedTemporaryFile(delete=False, suffix='.mp3') as tmpfile:
        tmpfile.write(response.content)
        return tmpfile.name

def merge_audio_files(audio_file_paths, output_file_path):
    """
    Merges multiple MP3 audio files into one MP3 file using the MoviePy library.

    Parameters:
    - audio_file_paths (list of str): A list of file paths for the MP3 files to be merged.
    - output_file_path (str): The file path where the merged MP3 file will be saved.

    """
    # Load each MP3 file as an AudioFileClip
    audio_clips = [AudioFileClip(audio_file) for audio_file in audio_file_paths]
    
    # Concatenate the audio clips into one final audio clip
    final_audio = concatenate_audioclips(audio_clips)
    
    # Write the final audio clip to the specified output file path
    final_audio.write_audiofile(output_file_path, codec='libmp3lame')
    
    # Close all the individual audio clips
    for clip in audio_clips:
        clip.close()

def main():
    """
    The main function of the script which performs the following steps:
    1. Parses the RSS feed from a specified URL.
    2. Converts the titles of feed entries to speech.
    3. Merges the generated speech audio files into a single file.
    4. Cleans up the temporary files.

    """
    # Define the feed URL for the RSS feed
    url = "https://www.lemonde.fr/sport/rss_full.xml"
    
    # Parse the feed using feedparser
    feed = feedparser.parse(url)

    # Extract titles from feed entries (could use entry.summary for more detailed text)
    texts_to_convert = [entry.title for entry in feed.entries]

    # Convert each title to speech and collect the file paths of the audio files
    generated_audio_file_paths = []
    for text in texts_to_convert:
        speech_file_path = text_to_speech(text)
        generated_audio_file_paths.append(speech_file_path)

    # Define the output file path for the merged audio file
    output_file_path = 'final_merged_audio.mp3'

    # Merge the generated audio files into a single file
    merge_audio_files(generated_audio_file_paths, output_file_path)

    # Clean up the temporary speech files created during the process
    for path in generated_audio_file_paths:
        os.remove(path)

    # Inform the user that the merged audio file has been created
    print(f"Merged audio file created at: {output_file_path}")

if __name__ == '__main__':
    main()  # Execute the main function if the script is run as the main module
```
